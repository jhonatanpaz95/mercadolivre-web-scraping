```python
import requests as req
from requests.exceptions import HTTPError
from bs4 import BeautifulSoup
```

```python
# link para download
link = 'https://www.mercadolivre.com.br/mais-vendidos/MLB1051#origin=home'

# função para requisição do link
def requisicao(url: str, headers:dict) -> str:

    try:
        resposta = req.get(url, headers=headers)
        resposta.raise_for_status()
    except HTTPError as exc:
        print(exc)
    except Exception  as e:
        print(e)
    else:
        return resposta.text
```

```python
#headers para parametro
HEADERS = {
    'User-Agent':'Mozila/5.0',
    'Accept-Language': 'en-US,en;q=0.5'
}

#dowload da pagina
extracao = requisicao(link, headers=HEADERS)
```

```python
#selecionando somente o html da pagina
pagina_mercadolivre = BeautifulSoup(extracao, 'html.parser')
```

```python

produtos = pagina_mercadolivre.find('div', {'class':'ui-search-layout--grid__grid'})
print(produtos)
```

```python
conteudo_extraido = []

for produto in produtos.find_all('a',{'class':'ui-recommendations-card__link'}):
  
  conteudo_extraido.append(produto)

print(conteudo_extraido[0])
```
